```python
from glob import glob
from pathlib import Path
from PIL import Image
from skimage.util import img_as_float
import cv2
import numpy as np
from pyaqua3ddev.image_processing.seathru import SeaThru
import torch
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"

device
```

```python
output_path = Path("./output")

png_files = [Path(f) for f in glob("../../data/seathru/**/*.png", recursive=True)]
files_tuple = [(f, f.parent.parent / "depth" / f"depth{f.stem}.tif", output_path / f.parent.parent.parent.stem / f.parent.parent.stem / "crutchfield_perry" / f"result_{f.stem}.png") for f in png_files]
files_tuple = [(f, t, o) for f, t, o in files_tuple if f.exists() and t.exists()]

files_tuple
```

```python
seathru = SeaThru(device)

for png_file, tif_file, output_file in files_tuple:
    output_file.parent.mkdir(parents=True, exist_ok=True)

    png = img_as_float(cv2.imread(png_file))

    tif_image = Image.open(tif_file)
    tif = np.array(tif_image)

    output = seathru.process(png, tif)

    cv2.imwrite(output_file, output)
```
